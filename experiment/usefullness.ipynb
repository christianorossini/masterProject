```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "last"
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
answers  =  pd.read_excel("answers.xls")

categories =  answers[~answers['coding'].isnull()]
categories = categories.loc[:,['dt_helped','coding']]

group_helped = categories.query('dt_helped==1').drop('dt_helped',axis=1)
group_helped.coding.value_counts()/sum(group_helped.coding.value_counts())
```

```python
data = group_helped.coding.value_counts().to_dict()
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots(figsize=(7, 4))

ax.barh(group_names, group_data)
ax.xaxis.grid(True)
ax.set_axisbelow(True) 
```

```python
group_nhelped = categories.query('dt_helped==0').drop('dt_helped',axis=1)

data = group_nhelped.coding.value_counts().to_dict()
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots(figsize=(6, 4))

# for i, v in enumerate(group_data):
#     ax.text(v - 0.38, i, str(v), color='white')

ax.barh(group_names, group_data, color="red")
ax.xaxis.grid(True)
ax.set_axisbelow(True) 
```

```python
categories.loc[categories.dt_helped==1,'useful'] = 'useful'
categories.loc[categories.dt_helped==0,'useful'] = 'useless'


categories.useful.value_counts().to_dict()


```
