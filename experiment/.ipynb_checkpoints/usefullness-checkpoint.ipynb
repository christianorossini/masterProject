```python
import numpy as np
import matplotlib.pyplot as plt


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(7, 4))
    ax.invert_yaxis()
    ax.xaxis.set_visible(True)
    #ax.set_xlim(0, np.sum(data, axis=1).max())
    ax.xaxis.grid(True)
    ax.set_axisbelow(True) 

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts,
                label=colname, color=color)
        xcenters = starts + widths / 2

#         r, g, b, _ = color
#         text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
#         for y, (x, c) in enumerate(zip(xcenters, widths)):
#             ax.text(x, y, str(int(c)), ha='center', va='center',
#                     color=text_color)
    ax.legend()

    return fig, ax

category_names = ['God Class', 'Long method', 'Long parameter list', 'Refused Bequest']
results = {
    'Question 1': [0, 0, 17, 32, 26],
    'Question 2': [26, 22, 29, 10, 13],
    'Question 3': [35, 37, 7, 2, 19],
    'Question 4': [32, 11, 9, 15, 33],
    'Question 5': [21, 29, 5, 5, 40],
    'Question 6': [8, 19, 5, 30, 38]
}

survey(results, category_names)
plt.show()
```

```python
answers  =  pd.read_excel("answers.xls")

categories =  answers[~answers['coding'].isnull()]
categories = categories.loc[:,['dt_helped','coding', 'cs']]

# group_helped = categories.query('dt_helped==1').drop('dt_helped',axis=1)
# group_helped.coding.value_counts()/sum(group_helped.coding.value_counts())

group_helped = categories.query('dt_helped==1').drop('dt_helped',axis=1).groupby(['coding','cs']).size().reset_index()
group_helped.columns=['categorie','cs','count']
group_helped = group_helped.pivot(columns="cs", values='count', index='categorie')
group_helped.fillna(0, inplace=True)
result = group_helped.T.to_dict(orient='list')

survey(result, category_names)
```

```python
data = group_helped.coding.value_counts().to_dict()
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots(figsize=(7, 4))

ax.barh(group_names, group_data)
ax.xaxis.grid(True)
ax.set_axisbelow(True) 
```

```python
group_nhelped = categories.query('dt_helped==0').drop('dt_helped',axis=1).groupby(['coding','cs']).size().reset_index()
group_nhelped.columns=['categorie','cs','count']
group_nhelped = group_nhelped.pivot(columns="cs", values='count', index='categorie')
group_nhelped.fillna(0, inplace=True)
result = group_nhelped.T.to_dict(orient='list')

survey(result, category_names)
```

```python
data = group_nhelped.coding.value_counts().to_dict()
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots(figsize=(6, 4))

# for i, v in enumerate(group_data):
#     ax.text(v - 0.38, i, str(v), color='white')

ax.barh(group_names, group_data, color="red")
ax.xaxis.grid(True)
ax.set_axisbelow(True) 
```
